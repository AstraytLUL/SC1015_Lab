```python
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt # we only need pyplot
sb.set() # set the default Seaborn style for graphics
```

```python
# importing data
NVDA_data = pd.read_csv('NVDA.csv')
AMD_data = pd.read_csv('AMD.csv')
ETH_data = pd.read_csv('ETH-USD.csv')

# merging data
NVDA_ETH_data = pd.merge(NVDA_data, ETH_data, on='Date')
AMD_ETH_data = pd.merge(AMD_data, ETH_data, on='Date')

# renaming data
NVDA_ETH_data.rename(columns = {'Close_x':'NVDA_Close', 'Open_x':'NVDA_Open', 'High_x':'NVDA_High', 'Low_x':'NVDA_Low', 'Adj Close_x':'NVDA_AdjClose', 'Volume_x':'NVDA_Volume', 'Close_y':'ETH_Close', 'Open_y':'ETH_Open', 'High_y':'ETH_High', 'Low_y':'ETH_Low', 'Adj Close_y':'ETH_AdjClose', 'Volume_y':'ETH_Volume'}, inplace = True)
AMD_ETH_data.rename(columns = {'Close_x':'AMD_Close', 'Open_x':'AMD_Open', 'High_x':'AMD_High', 'Low_x':'AMD_Low', 'Adj Close_x':'AMD_AdjClose', 'Volume_x':'AMD_Volume', 'Close_y':'ETH_Close', 'Open_y':'ETH_Open', 'High_y':'ETH_High', 'Low_y':'ETH_Low', 'Adj Close_y':'ETH_AdjClose', 'Volume_y':'ETH_Volume'}, inplace = True)

```

```python
f, axes = plt.subplots(3, 3, figsize=(24, 12))

# Plot the basic uni-variate figures for NVDA Close Price
sb.boxplot(data = NVDA_data['Close'], orient = "h", ax = axes[0,0])
sb.histplot(data = NVDA_data['Close'], ax = axes[0,1])
sb.violinplot(data = NVDA_data['Close'], orient = "h", ax = axes[0,2])

# Plot the basic uni-variate figures for AMD Close Price
sb.boxplot(data = AMD_data['Close'], orient = "h", ax = axes[1,0])
sb.histplot(data = AMD_data['Close'], ax = axes[1,1])
sb.violinplot(data = AMD_data['Close'], orient = "h", ax = axes[1,2])

# Plot the basic uni-variate figures for ETH Close Price
sb.boxplot(data = ETH_data['Close'], orient = "h", ax = axes[2,0])
sb.histplot(data = ETH_data['Close'], ax = axes[2,1])
sb.violinplot(data = ETH_data['Close'], orient = "h", ax = axes[2,2])
```

```python
# Set up the matplotlib figure
f = plt.figure(figsize = (24, 12))

# Create a joint plot
sb.jointplot(data=NVDA_ETH_data, x='NVDA_Open', y='ETH_Close', height=8, kind='scatter')

# Add title and axis labels
plt.title('NVDA Open Price vs. Ethereum Close Price')
plt.xlabel('NVDA Open Price')
plt.ylabel('Ethereum Close Price')

# Show the plot
plt.show()
```

```python
# Select only the 'NVDA_Open' and 'ETH_Close' columns
selected_columns = NVDA_ETH_data[['NVDA_Open', 'ETH_Close']]

# Calculate the correlation matrix
correlation_matrix = selected_columns.corr()

# Set up the matplotlib figure
plt.figure(figsize=(6, 4))

# Create a heatmap
sb.heatmap(correlation_matrix, vmin=-1, vmax=1, annot=True, fmt=".2f", cmap='coolwarm')

# Add title and adjust plot settings
plt.title('Correlation Heatmap: NVDA Open Price vs. Ethereum Close Price')

# Show the plot
plt.show()
```

```python
# Set up the matplotlib figure
f = plt.figure(figsize = (24, 12))

# Create a joint plot
sb.jointplot(data=AMD_ETH_data, x='AMD_Open', y='ETH_Close', height=8, kind='scatter')

# Add title and axis labels
plt.title('AMD Open Price vs. Ethereum Close Price')
plt.xlabel('AMD Open Price')
plt.ylabel('Ethereum Close Price')

# Show the plot
plt.show()
```

```python
# Select only the 'EthPriceNextDay' and 'BitPrice' columns
selected_columns = AMD_ETH_data[['AMD_Open', 'ETH_Close']]

# Calculate the correlation matrix
correlation_matrix = selected_columns.corr()

# Set up the matplotlib figure
plt.figure(figsize=(6, 4))

# Create a heatmap
sb.heatmap(correlation_matrix, vmin=-1, vmax=1, annot=True, fmt=".2f", cmap='coolwarm')

# Add title and adjust plot settings
plt.title('Correlation Heatmap: AMD Open Price vs. Ethereum Close Price')

# Show the plot
plt.show()
```

As we can see from the charts, the ETH closing price has very high correlation between both AMD and NVDA stock opening price.
